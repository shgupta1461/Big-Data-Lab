#### Performing above procedure on 10 pages of amazon.com and appending it to dataframe

```python
final = pd.DataFrame()

for j in range(1,11):
    url = "https://www.amazon.com/s?i=electronics-intl-ship&bbn=16225009011&rh=n%3A16225009011%2Cn%3A10048700011&page={}&qid=1706432446&ref=sr_pg_{}".format(j,j)
    page1 = scraper.get(url).text
    soup = BeautifulSoup(page1, 'html.parser')
    product = soup.find_all('div', class_="sg-col-inner")
    
    p_name = []
    price = []
    rating = []
    no_of_ratings = []
    
    for i in product[4:-2]:
    
        p_name.append(i.find('span', class_='a-size-base-plus a-color-base a-text-normal').text)
        rating.append(i.find('span', class_="a-icon-alt").text)
        no_of_ratings.append(i.find('span', class_="a-size-base s-underline-text").text)
        price.append(i.find('span', class_='a-offscreen').text)
    
    dict = {'Product Name': p_name, 'Rating': rating, 'Number of Ratings': no_of_ratings, 'Price': price} 
    df = pd.DataFrame(dict)
    
    final = final.append(df, ignore_index=True)
```

```python
final.shape
```

```python
final.to_csv("Wearable Technology Products on amazon.com")     #Saving file in CSV format
```
